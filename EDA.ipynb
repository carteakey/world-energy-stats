```python
import pandas as pd

# URL to the raw CSV file
url = 'https://raw.githubusercontent.com/Amarpreet3/DS8003---PROJECT/main/Dataset/World%20Energy%20Consumption.csv'

# Read the CSV file from the URL
WorldEnergyConsumption = pd.read_csv(url)

# Display the first few rows of the data
print(WorldEnergyConsumption)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = WorldEnergyConsumption.copy()
# Check for missing values
missing_values = df.isnull().sum()
missing_values
```

```python
# Check Info
df.info()
```

```python
# Summary statistics
summary_stats = df.describe()
summary_stats
```

```python
column_names = df.columns
print(column_names)
```

```python
# Example: Line plot of energy consumption change over years for a specific country
country_name = 'United States'  # Replace with the desired country name
country_data = df[df['country'] == country_name]
plt.figure(figsize=(12, 6))
plt.plot(country_data['year'], country_data['energy_cons_change_pct'], marker='o', linestyle='-')
plt.title(f'Energy Consumption Change Over Years in {country_name}')
plt.xlabel('Year')
plt.ylabel('Change Percentage')
plt.grid(True)
plt.show()
```
